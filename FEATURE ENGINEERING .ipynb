```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df1.drop(columns=['has_covid']),df1['has_covid'],test_size=0.2)

# train_test_split(pehle x vali value dete h , phir y vali value det h)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder
```

```python
transformer = ColumnTransformer(transformers = [
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['gender','city'])
],remainder = 'passthrough')
```

```python
transformer.fit_transform(x_train)
```

```python
transformer.transform(x_test)
```

<h2> HANDLING MIXED DATA </h2>

<b> lets suppose we have a column in which values with numerical and categorical data exist like s1 , b4 etc </b> <br>
<b> there is one more type that exist suppose column contains value like [1 2 3 A 5 6 A] here we have to first extract numerical value then categorical value  </b>

```python
import pandas as pd
dt = pd.read_csv('titanic.csv')
```

```python
dt.head(5)
```

```python
dt['number'].value_counts().plot.bar()
```

<b>first extract numerical part</b>

```python
dt['number_numerical'] = pd.to_numeric(dt['number'],errors='coerce',downcast='integer')
```

<b>then extract categorical data</b>

```python
dt['number_categorical'] = np.where(dt['number_numerical'].isnull(),dt['number'],np.nan)
dt.head()
```

```python
dt['Cabin'].unique()
```

```python
dt['Ticket'].unique()
```

<b> as we can see in cabin and in ticket column there is so much of unique data in file</b>

```python
dt['cabin_num'] = dt['Cabin'].str.extract('(\d+)') # captures numerical part 
dt['cabin_cat'] = dt['Cabin'].str[0] # captures categorical part 
```

```python
dt.head()
```

```python
dt['cabin_cat'].value_counts().plot.bar()
```
